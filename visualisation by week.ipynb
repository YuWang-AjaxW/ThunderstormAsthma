# 1. Initialisation

```python
#libraries
import pandas as pd
import re
```

```python
#read data
pollen = pd.read_csv("pollen_data_2009_2011.csv")
```

# 2. Data Wrangling

```python
# strip each column name
for each in pollen.columns:
    newCol = each.strip()
    pollen.rename(columns={each:newCol}, inplace=True)
```

```python
#check type of "first_date_of_weak"
pollen["first_date_of_week"].astype
```

```python
# edit "first_date_of_week"
for i in range(len(pollen)):
    obj = pollen["first_date_of_week"].iloc[i]
    date = re.split("T", obj)[0]
    pollen["first_date_of_week"].iloc[i] = date
```

```python
# add two new columns to pollen df
year = []
month = []

for i in range(len(pollen)):
    date = pollen["first_date_of_week"].iloc[i]
    y = re.split("-", date)[0]
    m = re.split("-", date)[1]
    year.append(y)
    month.append(m)
    
pollen["year"] = year
pollen["month"] = month
```

```python
#change type of "first_date_of_week" column
pollen["first_date_of_week"] = pd.to_datetime(pollen["first_date_of_week"])
```

```python
year2009 = [0] * 12
year2010 = [0] * 12
year2011 = [0] * 12


pollen1["month"] = pollen1["month"].astype(str).astype(int)


for y in pollen['year'].unique():
    count = 0
    subset = pollen1.loc[pollen1['year'] == y]
    for m in subset['month'].unique():
        count = subset.loc[subset['month'] == m, 'pollen_count'].sum()
        
        if y == '2009':
            year2009[m-1] = round(count,2)
        elif y == '2010':
            year2010[m-1] = round(count,2)
        elif y == '2011':
            year2011[m-1] = round(count,2)
```
