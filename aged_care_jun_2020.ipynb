```python
# convert "ABN" type from float to string
df = df.astype({"ABN": 'str'}) 
```

```python
df.dtypes
```

```python
# remove decimals of "ABN" column
for i in range(len(df)):
    abn = df['ABN'].iloc[i]
    if len(abn) == 13:
        abn = re.split("\.", abn)[0]
        df['ABN'].iloc[i] = abn
    elif len(abn) == 3:
        abn = None
```

```python
# correct address format
for i in range(len(df)):
    new_add = ""
    add = df['STREET_ST_ADDRESS'].iloc[i]
    for j in range(len(add)):
        current = add[j]
        pre = add[j-1]
        
        if pre == " ":
            new_add += add[j].upper()
        elif pre != " ":
            new_add += add[j].lower()
            
    df['STREET_ST_ADDRESS'].iloc[i] = new_add
```

```python
# correct suburb format
for i in range(len(df)):
    new_sub = ""
    #get each suburb name
    suburb = df['STREET_SUBURB'].iloc[i]
    
    #if there are have no space in the suburb name...
    if " " not in suburb:
        # capitalize the suburb name and replace the original value
        new_sub = suburb.capitalize()
        df['STREET_SUBURB'].iloc[i] = new_sub
        
    # elif there have spaces in the suburb name...
    elif " " in suburb:
        #convert all to lowercase
        suburb = suburb.lower()
        #make first letter to uppercase
        suburb = suburb.capitalize()
        #find spcace idx
        space_idx = []
        for j in range(len(suburb)):
            value = suburb[j]
            if value == " ":
                space_idx.append(j)
                
        #convert space index+1 to uppercase
        for n in space_idx:
            suburb = suburb[:n+1] + suburb[n+1].upper() + suburb[n+2:]
        #replace original values
        df['STREET_SUBURB'].iloc[i] = suburb
```

```python
# add new column 'address' to df
address = []
for i in range(len(df)):
    st = df['STREET_ST_ADDRESS'].iloc[i]
    sub = df['STREET_SUBURB'].iloc[i]
    postcode = df['STREET_PCODE'].iloc[i]
    add = st.strip() + " " + sub.strip() + " " + str(postcode).strip() + " VIC Australia"
    address.append(add)
```

```python
df['ADDRESS'] = address
```

```python
# remove duplicates
df = df.drop_duplicates(subset = ["ADDRESS"])
```
