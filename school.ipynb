# 3. Wrangling

```python
address = []
for i in range(len(df)):
    st = df['Address_Line_1'].iloc[i].strip()
    sub = df['Address_Town'].iloc[i].strip()
    state = df['Address_State'].iloc[i].strip()
    pc = df['Address_Postcode'].iloc[i]
    add = st + " " + sub + " " + state + " " + str(pc)
    address.append(add)
```

```python
df['Address'] = address
```

```python
name = []
add = []
town = []
lga = []
X = []
Y = []

for i in range(len(df)):
    n = df['School_Name'].iloc[i]
    name.append(n)
    a = df['Address'].iloc[i]
    add.append(a)
    t = df['Address_Town'].iloc[i]
    town.append(t)
    l = df['LGA_Name'].iloc[i]
    lga.append(l)
    x = df['X'].iloc[i]
    X.append(x)
    y = df['Y'].iloc[i]
    Y.append(y)
```

```python
schools = pd.DataFrame({'school_name':name, 'address':add, 'town':town, 'lga_name':lga, 'x':X, 'y':Y})
```

```python
schools.to_csv("schools_jun_20.csv", index = False)
```
